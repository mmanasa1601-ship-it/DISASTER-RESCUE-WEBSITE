```python
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI(title="Scam Detection & Honeypot System")
```

```python
class UserMessage(BaseModel):
    message: str
```

```python
import re
import math
def extract_nlp_features(text: str):
    text = text.lower()

    features = {
        "urgency": len(re.findall(r"urgent|immediately|act now", text)),
        "reward": len(re.findall(r"won|prize|lottery|reward", text)),
        "threat": len(re.findall(r"blocked|suspended|legal action", text)),
        "credentials": len(re.findall(r"otp|pin|password|cvv", text)),
        "money": len(re.findall(r"pay|transfer|upi|bank", text)),
        "links": len(re.findall(r"http|www|bit.ly|tinyurl", text))
    }
    return features
    def behavioral_anomaly_score(features: dict):
     score = 0

    score += features["urgency"] * 1.5
    score += features["reward"] * 1.2
    score += features["threat"] * 1.8
    score += features["credentials"] * 2.5
    score += features["money"] * 2.0
    score += features["links"] * 1.7

    return score
    def fraud_risk_engine(message: str):
     features = extract_nlp_features(message)
    anomaly_score = behavioral_anomaly_score(features)

    # Normalize risk score (0–100)
    risk_score = min(100, int((1 - math.exp(-anomaly_score)) * 100))

    if risk_score >= 70:
        label = "SCAM"
    elif risk_score >= 40:
        label = "SUSPICIOUS"
    else:
        label = "SAFE"

    return {
        "risk_score": risk_score,
        "classification": label,
        "features_detected": features
    }
    @app.post("/user/check-scam")
    def check_scam(data: UserMessage):
     result = fraud_risk_engine(data.message)

    if result["classification"] == "SCAM":
        return {
            "scam_detected": True,
            "risk_score": result["risk_score"],
            "reason": result["features_detected"],
            "action": "Alert police AI system"
        }
    else:
        return {
            "scam_detected": False,
            "risk_score": result["risk_score"],
            "status": result["classification"]
        }

```
