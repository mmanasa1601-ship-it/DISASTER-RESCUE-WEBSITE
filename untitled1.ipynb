```python
from fastapi import FastAPI
from pydantic import BaseModel

app = FastAPI(title="Scam Detection & Honeypot System")
```

```python
class UserMessage(BaseModel):
    message: str
```

```python
scam_keywords = [
    "urgent", "lottery", "won", "click",
    "verify", "blocked", "otp", "prize", "bank"
]

def detect_scam(message: str):
    score = 0
    for word in scam_keywords:
        if word in message.lower():
            score += 1
    return score >= 2
```

```python
@app.post("/check-scam")
def check_scam(data: UserMessage):
    is_scam = detect_scam(data.message)

    if is_scam:
        return {
            "scam_detected": True,
            "message": "⚠️ This is a SCAM. Do not reply or click any links."
        }
    else:
        return {
            "scam_detected": False,
            "message": "✅ This message looks safe."
        }
```
